```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from collections import Counter
from nltk.translate.meteor_score import meteor_score

def calculate_f1(ground_truth, prediction):
    """Calculates F1 score based on token overlap."""
    ground_truth_tokens = nltk.word_tokenize(ground_truth.lower())
    prediction_tokens = nltk.word_tokenize(prediction.lower())

    common_tokens = Counter(ground_truth_tokens) & Counter(prediction_tokens)
    num_common = sum(common_tokens.values())

    if num_common == 0:
        return 0.0

    precision = num_common / len(prediction_tokens)
    recall = num_common / len(ground_truth_tokens)

    if (precision + recall) == 0:
        return 0.0

    f1 = (2 * precision * recall) / (precision + recall)
    return f1

def calculate_meteor(ground_truth, prediction):
    """Calculates METEOR score based on semantic similarity and recall."""
    # meteor_score expects reference as a list of tokens, and hypothesis as a list of tokens.
    # It can also take a list of reference translations.
    reference = [nltk.word_tokenize(ground_truth.lower())]
    hypothesis = nltk.word_tokenize(prediction.lower())
    return meteor_score(reference, hypothesis)

# Apply the functions to the DataFrame
evaluation_df['F1'] = evaluation_df.apply(lambda row: calculate_f1(row['Ground_Truth_Answer'], row['Predicted_Answer']), axis=1)
evaluation_df['METEOR'] = evaluation_df.apply(lambda row: calculate_meteor(row['Ground_Truth_Answer'], row['Predicted_Answer']), axis=1)

print("Evaluation DataFrame with F1 and METEOR Scores:")
display(evaluation_df.head())

overall_f1 = evaluation_df['F1'].mean()
overall_meteor = evaluation_df['METEOR'].mean()

print(f"\nOverall F1 Score: {overall_f1:.3f}")
print(f"Overall METEOR Score: {overall_meteor:.3f}")
```
